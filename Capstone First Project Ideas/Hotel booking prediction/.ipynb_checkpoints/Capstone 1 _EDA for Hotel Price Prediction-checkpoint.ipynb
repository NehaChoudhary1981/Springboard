### Predicting Cancellations 
It would be great for the hotels to have a model to predict if a guest will actually arrive.


We can call the .corr() method on our Hotel data to look at all the correlations between our variables. The Pearson correlation coefficient shows us how close to a straight line the data-points fall, and is a number between -1 and 1.

```python
df.corr()
```

```python
#check the cancelation counts for Resort Hotel vs City Hotel

group_hotel_type = df.groupby(['hotel', 'is_canceled']).size().unstack(fill_value=0)
group_hotel_type.plot(kind='bar', stacked=True, figsize=(10,10),)
plt.title('Hotel types vs Booking Cancellation Status')
plt.xlabel('Hotel type', fontsize=12)
plt.xticks(rotation=360)
plt.ylabel('Count', fontsize=12)
```

The plot above and calculations highlight that the cancellation rate is higher for City Hotels (42%) compared to Resort Hotels(28%). The plot shows that city hotels have twice as many entries as resort hotel and hence the analysis is biased towards city hotels.

Total cancellations against bookings for both Hotels.

```python
c_df = df[df["hotel"] == "City Hotel"]
num_city_df = c_df['hotel'].count()
print('Total City Hotels:',num_city_df)

r_df = df[df["hotel"] == "Resort Hotel"]
num_resort_df = r_df['hotel'].count()
print('Total Resort Hotels:',num_resort_df)

c_can_df = c_df[c_df["is_canceled"] == 1]
c_can_count_df = c_can_df['is_canceled'].count()
c_can_per = c_can_count_df/num_city_df * 100
print('Total City Hotel Cancellations', c_can_count_df)

r_can_df = r_df[r_df["is_canceled"] == 1]
r_can_count_df = r_can_df['is_canceled'].count()
r_can_per = r_can_count_df/num_resort_df * 100
print('Total Resort Hotel Cancellations', r_can_count_df)

print('City Hotel Cancellations % ',c_can_per)
print('Resort Hotel Cancellations % ',r_can_per)
```

```python
#arrival_date_month exploration 
plt.figure(figsize=(15,10))
sns.countplot(x='arrival_date_month', data = df,order=pd.value_counts(df['arrival_date_month']).index,palette='YlOrBr_r')
plt.title('Arrival Month', weight='bold')
plt.xlabel('Month', fontsize=12)
plt.ylabel('Count', fontsize=12)
```

```python
plt.figure(figsize=(15,10))
sns.barplot(y='lead_time', x='arrival_date_month', hue='is_canceled',data=df,dodge=True,order=pd.value_counts(df['arrival_date_month']).index, palette='YlOrBr_r')
plt.title('Lead Time Vs Arrival_datemonth', weight='bold')
plt.xlabel('Lead Time', fontsize=12)
plt.ylabel('Arrival Month', fontsize=12)
```

This is slightly better. The plot indicates cancellation rates are higher when the lead times are long. 

```python
per_month_book_city = c_df.groupby("arrival_date_month")["hotel"].count()
month_can_city = c_df.groupby("arrival_date_month")["is_canceled"].sum()
#print('Monthly Bookings City Hotels:',per_month_book_city)
#print('Monthly Cancellations City Hotels:',month_can_city)
per_month_book_resort = r_df.groupby("arrival_date_month")["hotel"].count()
month_can_resort = r_df.groupby("arrival_date_month")["is_canceled"].sum()
#print('Monthly Bookings Resort Hotels:',per_month_book_resort)
#print('Monthly Cancellations Resort Hotels:',month_can_resort)

```

```python
city_cancel_df = pd.DataFrame({"Hotel": "City Hotel",
                            "Month": list(per_month_book_city.index),
                            "Bookings": list(per_month_book_city.values),
                            "Cancellations": list(month_can_city.values)})

resort_cancel_df = pd.DataFrame({"Hotel": "Resort Hotel",
                            "Month": list(per_month_book_resort.index),
                            "Bookings": list(per_month_book_resort.values),
                            "Cancellations": list(month_can_resort.values)})

Total_cancel_df = pd.concat([city_cancel_df, resort_cancel_df], ignore_index=True)
Total_cancel_df["cancel_percent"] = Total_cancel_df["Cancellations"] / Total_cancel_df["Bookings"] * 100
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x = "Month", y = "cancel_percent" , hue="Hotel",
            hue_order = ["City Hotel", "Resort Hotel"], data=Total_cancel_df,palette='YlOrBr_r')
plt.title("Cancelations per month", fontsize=16)
plt.xlabel("Month", fontsize=16)
plt.xticks(rotation=45)
plt.ylabel("Cancelations [%]", fontsize=16)
plt.legend(loc="upper right")
plt.show()
```
