8. Checking for NULL values in Dataset

```python
HotelBooking_data.isnull().values.any()
```

```python
#Count (using `.sum()`) the number of missing values (`.isnull()`) in each column of 
#HotelBooking_data as well as the percentages (using `.mean()` instead of `.sum()`).
#Order them (increasing or decreasing) using sort_values
#Call `pd.concat` to present these in a single table (DataFrame) with the helpful column names 'count' and '%'
missing = pd.concat([HotelBooking_data.isnull().sum(), 100 * HotelBooking_data.isnull().mean()], axis=1)
missing.columns=['Total Missing', '% Missing']
missing.sort_values(by='Total Missing')
```
